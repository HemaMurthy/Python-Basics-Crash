# HTML Parsing

```python
from  html.parser import HTMLParser
```

```python
metacount = 0
class myHTMLParser(HTMLParser):
    
    def handle_comment(self,data):
        print('Got Comment: ',data)
        pos = self.getpos()
        
        print('\tAt line ',pos[0],' position ',pos[1])
        
    def handle_starttag(self, tag, attrs):
        global metacount
        if tag == 'meta':
            metacount +=1
        print('Got Comment: ',tag)
        pos = self.getpos()
        
        print('\tAt line ',pos[0],' position ',pos[1])
        if attrs.__len__() >0 :
            for a in attrs:
                print(a[0], a[1])
        
    def handle_data(self, data):
        if(data.isspace()):
            return
        print('Got Comment: ',data)
        pos = self.getpos()
        
        print('\tAt line ',pos[0],' position ',pos[1])
        
        
```

```python
parser = myHTMLParser()

f = open("samplehtml.html")
if f.mode == 'r':
    print('Successful Read!')
    content = f.read()
    parser.feed(content)
    
```
